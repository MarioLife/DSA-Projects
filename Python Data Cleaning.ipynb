```python
diamonds_db = pd.read_csv('diamonds.csv')
import matplotlib.pyplot as plt
```

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,figsize=(20,10))

fig.suptitle('Diamonds - Univariate Analysis', size=25)

colors = diamonds_db['color'].value_counts()
colors.plot(kind = 'bar', ax = ax1, rot=True, color = 'red', title='Color')

cuts = diamonds_db['cut'].value_counts()
cuts.plot(kind = 'bar', ax = ax2, rot = True, color = 'blue', title='Cut')

price = diamonds_db['price']
price.plot(kind = 'box', ax = ax3, title = 'Price')

clarity = diamonds_db['clarity'].value_counts()
clarity.plot(kind = 'barh', ax = ax4, color = 'green', title = 'Clarity')
```
